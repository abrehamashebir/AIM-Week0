DISPLAY USING BOXPLOT

```python
plt.figure(figsize=(10,19))
sns.boxplot(data=df, x='RH',y='Timestamp' )
plt.title('GHI by Time')
plt.show()
```

```python


# Convert date column to datetime format if not already done
df['date'] = pd.to_datetime(df['Timestamp'])

# Set date as the index
df.set_index('date', inplace=True)

# Plotting GHI, DNI, DHI, and Tamb
plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
plt.plot(df.index, df['GHI'], label='GHI', color='orange')
plt.plot(df.index, df['DNI'], label='DNI', color='blue')
plt.plot(df.index, df['DHI'], label='DHI', color='green')
plt.title('Solar Radiation Over Time')
plt.xlabel('Date')
plt.ylabel('Solar Radiation (W/m²)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(df.index, df['Tamb'], label='Tamb', color='red')
plt.title('Ambient Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
correlation_columns = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']
correlation_matrix = df[correlation_columns].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
scatter_matrix_columns = ['GHI', 'DNI', 'DHI', 'WS', 'WSgust', 'WD']

# Create scatter matrix
pd.plotting.scatter_matrix(df[scatter_matrix_columns], figsize=(12, 12), alpha=0.5)
plt.suptitle('Scatter Matrix of Wind Conditions and Solar Irradiance', y=1.02)
plt.show()
```

```python
# Extract month and year
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['month'] = df['Timestamp'].dt.month

# Group by month and calculate the average solar radiation
monthly_avg = df.groupby('month')['solar_radiation'].mean()

# Plot monthly average solar radiation
plt.figure(figsize=(10, 5))
monthly_avg.plot(kind='bar', color='skyblue')
plt.title('Average Solar Radiation by Month')
plt.xlabel('Month')
plt.ylabel('Average Solar Radiation (W/m²)')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

```python
# Heatmap of correlations
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
